```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 624

varieties = ['Rosemary', 'Chamomile', 'Lavender', 'Mint', 'Eucalyptus']
seasons   = ['Summer', 'Winter', 'Spring', 'Monsoon']
altitudes = ['High (>2500m)', 'Mid (1500-2500m)', 'Low (<1500m)']

df = pd.DataFrame({
    'batch_id'                 : [f'B{str(i).zfill(3)}' for i in range(1, n+1)],
    'plant_variety'            : np.random.choice(varieties, n),
    'harvest_season'           : np.random.choice(seasons, n),
    'altitude_zone'            : np.random.choice(altitudes, n),
    'input_kg'                 : np.round(np.random.uniform(30, 70, n), 1),
    'distillation_duration_hrs': np.round(np.random.uniform(3, 6, n), 1),
    'temperature_c'            : np.round(np.random.uniform(70, 90, n), 1),
    'output_ml'                : np.round(np.random.uniform(1000, 2500, n), 0),
})

df['yield_efficiency_pct'] = np.round((df['output_ml'] / (df['input_kg'] * 1000)) * 100, 2)
df['quality_grade']        = pd.cut(df['yield_efficiency_pct'],
                                     bins=[0, 3, 3.5, 4, 100],
                                     labels=['D', 'C', 'B', 'A'])

print("Dataset ready!", df.shape)
```

```python
# CHART 1: LINE CHART - Yield Efficiency Trend Over Batches
plt.figure(figsize=(10,6))
df_sorted = df.sort_values('batch_id').reset_index()
rolling_avg = df_sorted['yield_efficiency_pct'].rolling(window=20).mean()

plt.plot(df_sorted.index, df_sorted['yield_efficiency_pct'], color='#a8d5ba', alpha=0.4, label='Raw Yield (per batch)')
plt.plot(df_sorted.index, rolling_avg, color='#1b5e20', linewidth=2.5, label='20-Batch Rolling Average')

plt.title('Yield Efficiency Trend Across Production Batches', fontsize=13, fontweight='bold')
plt.xlabel('Batch Sequence (chronological order)', fontsize=10)
plt.ylabel('Yield Efficiency (%)', fontsize=10)
plt.legend()
plt.tight_layout()
plt.show()
```

**Chart 1: Line Chart**

Business Question: How does extraction yield efficiency trend across production batches over time?

Why this chart: A line chart is best suited for showing trends over a sequence — here it reveals whether yield is improving, declining, or staying stable as more batches are processed, which a bar or pie chart cannot show.

```python
# CHART 2: BAR CHART - Average Yield Efficiency by Plant Variety
plt.figure(figsize=(9,6))
avg_yield = df.groupby('plant_variety')['yield_efficiency_pct'].mean().sort_values(ascending=False)

colors = ['#1b5e20', '#2e7d32', '#43a047', '#66bb6a', '#a5d6a7']
plt.bar(avg_yield.index, avg_yield.values, color=colors)

plt.title('Average Yield Efficiency by Plant Variety', fontsize=13, fontweight='bold')
plt.xlabel('Plant Variety', fontsize=10)
plt.ylabel('Average Yield Efficiency (%)', fontsize=10)
for i, v in enumerate(avg_yield.values):
    plt.text(i, v + 0.05, f'{v:.2f}%', ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

**Chart 2: Bar Chart**

Business Question: Which plant variety produces the highest average yield efficiency?

Why this chart: A bar chart is ideal for comparing a metric across discrete categories — here it clearly ranks the 5 plant varieties by performance, making it easy to spot which one is most efficient to process.

```python
# CHART 3: PIE CHART - Quality Grade Distribution
plt.figure(figsize=(8,8))
grade_counts = df['quality_grade'].value_counts().sort_index()

colors_pie = ['#c62828', '#f9a825', '#66bb6a', '#1b5e20']
plt.pie(grade_counts.values, labels=grade_counts.index, autopct='%1.1f%%',
        colors=colors_pie, startangle=90, textprops={'fontsize': 11})

plt.title('Distribution of Quality Grades Across All Batches', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Chart 3: Pie Chart**

Business Question: What proportion of total batches fall into each quality grade?

Why this chart: A pie chart works well here because there are only 4 quality grade categories, and the goal is to show composition (share of whole) rather than precise comparison — perfect for a quick visual of overall production quality.

```python
# CHART 4: SCATTER PLOT - Input Quantity vs Output Volume
plt.figure(figsize=(9,6))
scatter = plt.scatter(df['input_kg'], df['output_ml'],
                       c=df['yield_efficiency_pct'], cmap='Greens',
                       alpha=0.7, edgecolors='black', linewidth=0.3, s=40)

plt.colorbar(scatter, label='Yield Efficiency (%)')
plt.title('Input Quantity vs Output Volume (coloured by Yield Efficiency)', fontsize=13, fontweight='bold')
plt.xlabel('Input Quantity (kg)', fontsize=10)
plt.ylabel('Output Volume (ml)', fontsize=10)
plt.tight_layout()
plt.show()
```

**Chart 4: Scatter Plot**

Business Question: Is there a relationship between the amount of raw material used and the oil output produced?

Why this chart: A scatter plot is the right choice for examining the relationship between two continuous numeric variables — here it reveals whether higher input consistently leads to higher output, or if other factors are driving yield instead.

```python
# CHART 5: HEATMAP - Correlation Matrix
plt.figure(figsize=(9,7))
corr_cols = ['input_kg', 'distillation_duration_hrs', 'temperature_c', 'output_ml', 'yield_efficiency_pct']
corr_matrix = df[corr_cols].corr().round(2)

sns.heatmap(corr_matrix, annot=True, cmap='Greens', fmt='.2f',
            linewidths=0.5, cbar_kws={'label': 'Correlation Coefficient'})

plt.title('Correlation Matrix — Batch Yield Variables', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Chart 5: Heatmap**

Business Question: Which variables are most strongly correlated with yield efficiency, and could any be used to predict it?

Why this chart: A heatmap is best for visualising correlation between multiple numeric variables at once — it instantly highlights which factors (like input_kg) have the strongest relationship with yield efficiency, something a single scatter plot can't show across all variables simultaneously.
